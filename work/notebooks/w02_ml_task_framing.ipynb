## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

knowing why AI reffered a content is currently like a blackbox. Since, it's still in early stages, there is not definite information about what works and what doesn't work. Moreover, there are several AI models (gemini, chatgpt, claude and so on) that can refered. This means rules that works for a model might not work for the others. So, setting a rule can be a complicated and ambiguous.

Analysing a data, learning the patterns, deriving and picking the right features that would work, and feeding it to an AI model is like the right thing to do. If done well, our selected model can generalize to unseen data and do better than writing rules.
